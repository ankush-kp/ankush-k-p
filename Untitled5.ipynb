```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv('Heart.csv')
df.head()
```

```python
pd.set_option("display.float", "{:.2f}".format)
df.describe()
```

```python
df.AHD.value_counts().plot(kind="bar", color=["salmon", "lightblue"])
```

```python
df.isna().sum()
```

```python
categorical_val = []
continous_val = []
for column in df.columns:
  print('========================================================================')
  print(f'{column} : {df[column].unique()}')
  if len(df[column].unique()) <= 10:
    categorical_val.append(column)
  else:
    continous_val.append(column)
```

```python
plt.figure(figsize=(15, 15))

for i, column in enumerate(categorical_val, 1):
  plt.subplot(3, 3, i)
  df[df['AHD'] == 0][column].hist(bins=35, color='blue', label='Have Heart Disease = NO', alpha=0.6)
  df[df['AHD'] == 1][column].hist(bins=35, color='red', label='Have Heart Disease = YES', alpha=0.6)
  plt.legend()
  plt.xlabel(column)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))

for i, column in enumerate(continous_val, 1):
  plt.subplot(3, 3, i)
  df[df['AHD'] == 0][column].hist(bins=35, color='blue', label='Have Heart Disease = NO', alpha=0.6)
  df[df['AHD'] == 1][column].hist(bins=35, color='red', label='Have Heart Disease = YES', alpha=0.6)
  plt.legend()
  plt.xlabel(column)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

plt.scatter(df.Age[df.AHD == 1],
            df.MaxHR[df.AHD == 1],
            c="salmon")

plt.scatter(df.Age[df.AHD == 0],
            df.MaxHR[df.AHD == 0],
            c='lightblue')

plt.title("Heart Disease in function of Age and Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(["Disease", "No Disease"])

plt.show()
```

```python
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(15, 15))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
